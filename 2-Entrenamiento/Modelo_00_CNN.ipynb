```python
import pandas as pd
import numpy as np
import torch
import torch.nn as nn
import torch.nn.functional as F
from torch.utils.data import Dataset, DataLoader
from sklearn.model_selection import train_test_split
from sklearn.metrics import roc_auc_score, accuracy_score, classification_report
from sklearn.preprocessing import StandardScaler

# Load your data (assuming df is your DataFrame)
# df = pd.read_csv('your_data.csv')  # Uncomment if loading from file

# Create feature names - all columns except time, segment, and kind
features = [col for col in df.columns if col not in ["time", "segment", "kind"]]
print(f"Number of features: {len(features)}")

# Prepare features and target
X = df[features].fillna(0).astype(np.float32)
y = (df["kind"] == "word").astype(int)  # Convert to int instead of float32

# Check class distribution
print(f"Class distribution: {np.bincount(y)}")
print(f"Percentage of 'word' class: {y.mean():.2%}")

# Standardize features (important for neural networks)
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)
X_scaled = pd.DataFrame(X_scaled, columns=features, index=X.index)

# Split data with stratification
X_train, X_val, y_train, y_val = train_test_split(
    X_scaled, y, test_size=0.2, stratify=y, random_state=42
)

class SpeechDataset(Dataset):
    def __init__(self, X, y):
        self.X = torch.tensor(X.values, dtype=torch.float32)
        self.y = torch.tensor(y.values if hasattr(y, 'values') else y, dtype=torch.float32)
    
    def __len__(self):
        return len(self.X)
    
    def __getitem__(self, idx):
        return self.X[idx], self.y[idx]

# Create data loaders
train_dataset = SpeechDataset(X_train, y_train)
val_dataset = SpeechDataset(X_val, y_val)

train_loader = DataLoader(train_dataset, batch_size=128, shuffle=True)
val_loader = DataLoader(val_dataset, batch_size=128, shuffle=False)

class MLPWithAttention(nn.Module):
    def __init__(self, input_dim, hidden_dims=[256, 128], dropout_rate=0.3):
        super().__init__()
        
        # Attention mechanism
        self.attention = nn.Sequential(
            nn.Linear(input_dim, input_dim // 2),
            nn.ReLU(),
            nn.Linear(input_dim // 2, input_dim),
            nn.Sigmoid()
        )
        
        # Main network with dropout for regularization
        self.fc1 = nn.Linear(input_dim, hidden_dims[0])
        self.dropout1 = nn.Dropout(dropout_rate)
        self.fc2 = nn.Linear(hidden_dims[0], hidden_dims[1])
        self.dropout2 = nn.Dropout(dropout_rate)
        self.out = nn.Linear(hidden_dims[1], 1)
        
        # Initialize weights
        self._init_weights()
    
    def _init_weights(self):
        for m in self.modules():
            if isinstance(m, nn.Linear):
                nn.init.xavier_uniform_(m.weight)
                nn.init.constant_(m.bias, 0)
    
    def forward(self, x):
        # Apply attention
        attn_weights = self.attention(x)
        x = x * attn_weights
        
        # Forward pass through network
        x = F.relu(self.fc1(x))
        x = self.dropout1(x)
        x = F.relu(self.fc2(x))
        x = self.dropout2(x)
        
        return self.out(x).squeeze(1)

# Set device and model
device = torch.device("cuda" if torch.cuda.is_available() else "cpu")
print(f"Using device: {device}")

model = MLPWithAttention(input_dim=X_scaled.shape[1]).to(device)
optimizer = torch.optim.Adam(model.parameters(), lr=1e-3, weight_decay=1e-5)
criterion = nn.BCEWithLogitsLoss()

# Learning rate scheduler
scheduler = torch.optim.lr_scheduler.ReduceLROnPlateau(
    optimizer, mode='max', factor=0.5, patience=10, verbose=True
)

# Training loop with improved monitoring
best_auc = 0
patience_counter = 0
max_patience = 20

for epoch in range(200):
    # Training
    model.train()
    train_loss = 0
    train_correct = 0
    train_total = 0
    
    for xb, yb in train_loader:
        xb, yb = xb.to(device), yb.to(device)
        
        # Forward pass
        logits = model(xb)
        loss = criterion(logits, yb)
        
        # Backward pass
        optimizer.zero_grad()
        loss.backward()
        
        # Gradient clipping to prevent exploding gradients
        torch.nn.utils.clip_grad_norm_(model.parameters(), max_norm=1.0)
        
        optimizer.step()
        
        # Track metrics
        train_loss += loss.item()
        preds = (torch.sigmoid(logits) > 0.5).float()
        train_correct += (preds == yb).sum().item()
        train_total += yb.size(0)
    
    # Validation
    model.eval()
    val_loss = 0
    val_preds = []
    val_targets = []
    
    with torch.no_grad():
        for xb, yb in val_loader:
            xb, yb = xb.to(device), yb.to(device)
            
            logits = model(xb)
            loss = criterion(logits, yb)
            val_loss += loss.item()
            
            probs = torch.sigmoid(logits)
            val_preds.extend(probs.cpu().numpy())
            val_targets.extend(yb.cpu().numpy())
    
    # Calculate metrics
    val_preds = np.array(val_preds)
    val_targets = np.array(val_targets)
    
    auc = roc_auc_score(val_targets, val_preds)
    val_accuracy = accuracy_score(val_targets, val_preds > 0.5)
    train_accuracy = train_correct / train_total
    
    # Update learning rate
    scheduler.step(auc)
    
    # Print progress every 10 epochs
    if (epoch + 1) % 10 == 0:
        print(f"Epoch {epoch+1:3d} | "
              f"Train Loss: {train_loss/len(train_loader):.4f} | "
              f"Train Acc: {train_accuracy:.4f} | "
              f"Val Loss: {val_loss/len(val_loader):.4f} | "
              f"Val Acc: {val_accuracy:.4f} | "
              f"Val AUC: {auc:.4f}")
    
    # Early stopping
    if auc > best_auc:
        best_auc = auc
        patience_counter = 0
        # Save best model
        torch.save(model.state_dict(), 'best_model.pth')
    else:
        patience_counter += 1
        if patience_counter >= max_patience:
            print(f"Early stopping at epoch {epoch+1}")
            break

# Load best model for final evaluation
model.load_state_dict(torch.load('best_model.pth'))
model.eval()

# Final evaluation
final_preds = []
final_targets = []

with torch.no_grad():
    for xb, yb in val_loader:
        xb, yb = xb.to(device), yb.to(device)
        logits = model(xb)
        probs = torch.sigmoid(logits)
        final_preds.extend(probs.cpu().numpy())
        final_targets.extend(yb.cpu().numpy())

final_preds = np.array(final_preds)
final_targets = np.array(final_targets)

# Print final results
print("\n" + "="*50)
print("FINAL RESULTS")
print("="*50)
print(f"Best Validation AUC: {roc_auc_score(final_targets, final_preds):.4f}")
print(f"Validation Accuracy: {accuracy_score(final_targets, final_preds > 0.5):.4f}")
print("\nClassification Report:")
print(classification_report(final_targets, final_preds > 0.5, 
                          target_names=['silence', 'word']))

```
